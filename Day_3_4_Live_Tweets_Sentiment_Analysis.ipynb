```
import pandas as pd
df=pd.DataFrame()
df['Tweets']=tweets
```

```
score = []
for i in range(len(public_tweets)):
  review = TextBlob(tweets[i]).sentiment.polarity
  if review>0:
    score.append('pos')
  else:
    score.append('neg')
df['Score'] = score
```

```
df
```

```
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```
polarity_scores = []
for i in range(0,len(public_tweets)):
  p = sia.polarity_scores(tweets[i])
  polarity_scores.append(p)
```

```
scores = []
for i in range(0,len(polarity_scores)):
  c = polarity_scores[i]['compound']
  if c>0:
    scores.append('pos')
  else:
    scores.append('neg')
```

```
df['Score_nltk'] = scores
```

```
df
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df["Score"],df["Score_nltk"])
cm
```

```
from sklearn.metrics import accuracy_score
accu = accuracy_score(df["Score"],df["Score_nltk"])
accu
```
